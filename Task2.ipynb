```python
!python evaluate_train_rankings.py --re_ranking e5mistral_claims.json
```

```python
import json

# Load prediction files
with open("claims_mistral.json", "r") as f1:
    pred1 = json.load(f1)
with open("e5mistral_claims.json", "r") as f2:
    pred2 = json.load(f2)

fused_predictions = {}

for query_fan in pred1:
    docs1 = pred1.get(query_fan, [])
    docs2 = pred2.get(query_fan, [])

    # Build rank-based score dicts
    score1 = {doc: 1 / (i + 1) for i, doc in enumerate(docs1)}
    score2 = {doc: 1 / (i + 1) for i, doc in enumerate(docs2)}

    # Union of all candidate docs
    all_docs = set(docs1) | set(docs2)

    # Average scores
    fused_scores = {
    doc: 0.5 * score1.get(doc, 0) + 0.5 * score2.get(doc, 0)
    for doc in all_docs
    }

    # Sort by fused score descending
    sorted_docs = sorted(fused_scores.items(), key=lambda x: x[1], reverse=True)
    fused_predictions[query_fan] = [doc for doc, _ in sorted_docs]

# Save fused prediction
with open("prediction_fused_linq_e5.json", "w") as f:
    json.dump(fused_predictions, f, indent=2)

print(" Fused predictions saved to prediction_fused_linq_e5.json")
```
